# Reversed Range

range() funksiyasini reverse qilishning 2 metodi bor.



*   negative **step** dan foydalanish

*   reversed() funksiyasidan foydalanish.

python-reverse-range.jpg


```python
#Negative step indexing

# start = 5
# stop = -1
# step = -1
for i in range(5, -1, -1):
    print(i)
```

```python
#Using reversed() function
for i in reversed(range(10, 21)):
    print(i, end=' ')
```

```python
#get type of reversed range

print(type(range(0, 5)))
# Output;  <class 'range'>

print(type(reversed(range(0, 5))))
# Output: <class 'range_iterator'>
```
